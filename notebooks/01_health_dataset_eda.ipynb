```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Load the dataset
ds = load_breast_cancer(as_frame=True)
df = ds.frame.copy()
df.rename(columns={'target': 'diagnosis'}, inplace=True)

# Display first few rows
df.head()
```

```python
# Basic information about the data
df.info()
df.describe().T.head(10)
df['diagnosis'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for cleaner visuals
sns.set(style="whitegrid")

# Plot the distribution of diagnosis (Malignant vs Benign)
sns.countplot(x='diagnosis', data=df, palette='coolwarm')
plt.title("Diagnosis Distribution")
plt.show()

# Plot correlation heatmap for first 10 features
plt.figure(figsize=(10, 8))
sns.heatmap(df.iloc[:, :10].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (First 10 Features)")
plt.show()
```
